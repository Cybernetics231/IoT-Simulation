```python
#load class
import numpy as np
import pandas as pd 

import os
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.preprocessing import StandardScaler #module for perform scaling
```

```python
from sklearn.model_selection import train_test_split #module for splitting datatset
from sklearn import metrics #module for evaluating performance
```

```python
#load your data
df = pd.read_csv("better_data.csv") #change the name accordingly
df.head() # prints top 5 rows from the datatset to check data is load or not
```

```python
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
df = df.drop_duplicates()
```

```python
df.head()
```

```python
# prepare features
x = df.drop(['Label'],axis=1) #remove class or label
y = df['Label'] #load label
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.2) #split datatset. Here ratio is 80:20. Change accordingly
# Scale the data using standardization
scaler = StandardScaler()

x_train = scaler.fit_transform(x_train) #scale training set
x_test = scaler.transform(x_test) #scale test set

```

```python
from sklearn.linear_model import LogisticRegression

model1 = LogisticRegression()
model1.fit(x_train, y_train)
predict1 = model1.predict(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model2 = RandomForestClassifier(n_estimators=100, random_state=42)
model2.fit(x_train, y_train)
predict2 = model2.predict(x_test)
```

```python
from sklearn.svm import SVC
model3 = SVC()
model3.fit(x_train, y_train)
predict3 = model3.predict(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model4 = KNeighborsClassifier(n_neighbors=5)
model4.fit(x_train, y_train)
predict4 = model4.predict(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
model5 = DecisionTreeClassifier(random_state=42)
model5.fit(x_train, y_train)
predict5 = model5.predict(x_test)
```

```python
# Model Checks
check_df = pd.DataFrame({
    'Actual': y_test, 
    'Predicted_LR': predict1, 
    'Predicted_RF': predict2,
    'Predicted_SVM': predict3,
    'Predicted_KNN': predict4,
    'Predicted_DT': predict5
})
print(check_df.head(30))
```

```python
print("Accuracy LR:", metrics.accuracy_score(y_test, predict1))
print("Accuracy RF:", metrics.accuracy_score(y_test, predict2))
print("Accuracy SVM:", metrics.accuracy_score(y_test, predict3))
print("Accuracy KNN:", metrics.accuracy_score(y_test, predict4))
print("Accuracy Decision Tree:", metrics.accuracy_score(y_test, predict5))
```
